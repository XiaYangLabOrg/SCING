```python
import leidenalg        # Version 0.8.4
import igraph as ig
import numpy as np
import pandas as pd
from collections import Counter
import scanpy as sc
#import matplotlib.pyplot as plt

#import seaborn as sns
from pyscenic.genesig import GeneSignature
from pyscenic.aucell import aucell, derive_auc_threshold, create_rankings
from sklearn.decomposition import PCA
```

```python
adata = sc.read('../data/microglia.h5ad')
network = pd.read_csv('../data/micro.network.merged.csv.gz')
resolution_parameter = 0.0011
```

```python
adata
```

```python
# create and partition graph
G = ig.Graph.TupleList([tuple(x) for x in network.values],
                       directed = True)

partition = leidenalg.find_partition(G, leidenalg.CPMVertexPartition, resolution_parameter = resolution_parameter);
groups = np.array(partition.membership)
```

```python
# organizing gene membership dataframe
genes = G.vs['name']
gene_membership = pd.DataFrame(np.array([genes,groups]).T, columns = ['genes','cluster_membership'])
gene_membership = gene_membership.sort_values('genes')
```

```python
# remove modules with 2 or fewer genes
group_count = Counter(gene_membership.cluster_membership)
groups_to_keep = []
for g in group_count:
    if group_count[g] > 2:
        groups_to_keep.append(g)
groups_to_keep = np.array(groups_to_keep)
gene_membership = gene_membership.loc[np.isin(gene_membership.cluster_membership,
                                              groups_to_keep)]
```

```python
gene_membership.cluster_membership = gene_membership.cluster_membership.astype(str)
```

```python
# save gene memberships for pathway enrichment analysis
gene_membership.to_csv('../intermediate_data/gene.membership.csv.gz',
                      index=None)
```

```python
# test.gmt will need to be changed if doing this many times
# in parallel
with open('test.gmt','w') as f:
    for tup in gene_membership.groupby(by='cluster_membership'):
        f.write(str(tup[0]+'\t'))
        for g in tup[1].genes:
            f.write(g+'\t')
        f.write('\n')
        
signatures = GeneSignature.from_gmt('test.gmt',
                                    field_separator='\t',
                                   gene_separator='\t')

adata = adata[:,gene_membership.genes.values.ravel()]
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
dge = pd.DataFrame(adata.X.T.todense())
dge.index = adata.var.index
```

```python
# 0.05 can be modified 
percentiles = pd.Series(np.count_nonzero(dge.T, axis=1)).quantile(np.arange(0,1,0.001)) / dge.T.shape[1]
auc_thresh = percentiles[percentiles > 0.05].index[0]
```

```python
# get the AUCell scores for each cell across modules
# change num workers to number of cores available
aucs_mtx = aucell(dge.T, signatures,
                  auc_threshold=auc_thresh,
                  num_workers=12)
```

```python
# this is the cells x modules matrix of AUCell scores
aucs_mtx.head()
```

```python
# PCA on AUCell scores
pca = PCA(n_components=20).fit(aucs_mtx.values)
embedded_auc_mtx = pca.transform(aucs_mtx.values)
```

```python
# UMAP on AUCell + PCA
adata.obsm['X_network_pca'] = embedded_auc_mtx
sc.pp.neighbors(adata, n_neighbors=20, use_rep='X_network_pca')
sc.tl.umap(adata, min_dist=.1, spread=1)
```

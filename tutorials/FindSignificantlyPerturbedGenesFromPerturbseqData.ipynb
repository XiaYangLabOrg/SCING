# Finding significantly perturbed genes downstream of perturbed gene
This takes about a full day to run all the way through. I recommend running it in a shell script separately for a day to save all of the intermediate files and then you can run through it to see the output at each step without waiting for every step.

```python
# load data
adata = sc.read('../data/papalexi.anndata.h5ad')
```

```python
# which gene to find the significantly perturbed genes for
perturbed_gene = 'IFNGR2'

# there are multiple probes for each gene
g_indices = []
guide_names = []
for i,g in enumerate(np.unique(adata.obs.guide_ID)):
    if perturbed_gene in g:
        g_indices.append(i)
        guide_names.append(g)


# resolution for subclustering of cells
resolution = 1.2

# number of permutations to determine significance of a probes perturbation effect
npermutations = 100

# if you want to ensure overwriting of the intermediate files
overwrite = False
```

```python
adata = adata[:,np.mean(adata.X != 0 ,axis=0) > 0.05]
```

```python
adata
```

```python
# check to make sure the perturbed gene is in the data
perturbed_gene in adata.var.index
```

```python
# build a design matrix from the probes
uniq_perturbs = np.unique(adata.obs.guide_ID)
design_matrix = np.zeros((adata.shape[0],
                         len(uniq_perturbs)))
design_matrix = pd.DataFrame(design_matrix)
design_matrix.index = adata.obs.index
design_matrix.columns = uniq_perturbs
```

```python
for tup in zip(adata.obs.index,
               adata.obs.guide_ID):
    design_matrix.loc[tup[0],tup[1]] = 1
```

```python
# prepare the data for subclustering
sc.pp.calculate_qc_metrics(adata, percent_top=None, log1p=False, inplace=True)
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
adata.raw = adata
adata = adata[:, adata.var.highly_variable]

sc.pp.scale(adata, max_value=10)
sc.tl.pca(adata, svd_solver='arpack',n_comps=50)

sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata)

```

```python
# check where the control probes are in the cells
control_indices = np.where(adata.obs.gene == 'NT')[0]
```

```python
# separate out the control cells (cells without real perturbation)
control_cells = adata[control_indices]
sc.pp.neighbors(control_cells, n_neighbors=10, n_pcs=20)
sc.tl.umap(control_cells)
```

```python
# cluster only the cells that are control
sc.tl.leiden(control_cells,
            resolution=resolution)

uniq_leiden = np.unique(control_cells.obs.leiden)


sc.pl.umap(control_cells, color=['leiden',
                                 'n_genes_by_counts'])
```

```python
# examine the predicted state of the cells with real perturbations
clf=SVC(probability=True,
        C=1).fit(control_cells.obsm['X_pca'],
                            control_cells.obs.leiden)

predicted = clf.predict(adata.obsm['X_pca'])

adata.obs.loc[:,'leiden'] = predicted

sc.pl.umap(adata, color='leiden', legend_loc='on data')
```

```python
# predict the probability of each cell being in a given state
# these are used as covariates in the data
predicted_prob = clf.predict_proba(adata.obsm['X_pca'])
leiden_probs = ['leiden_'+i for i in uniq_leiden]
adata.obs[leiden_probs] = predicted_prob
```

```python
# plot the prediction probabilities for each state
sc.pl.umap(adata, color=leiden_probs,
          cmap='inferno',
          vmax=1,
          vmin=0)
```

```python
design_matrix[leiden_probs] = adata.obs[leiden_probs]
```

```python
design_matrix['total_counts'] = adata.obs.total_counts
```

```python
# runs the elastic net regression and get the regression coefficients
def run_reg(x, y, gene_names, uniq_perturbs, regr):
    regr.fit(x, y)
    coefs = pd.DataFrame(regr.coef_)
    coefs.index = gene_names
    coefs.columns = uniq_perturbs
    
    return coefs
```

```python
# add in total counts as a covariate
uniq_perturbs = list(uniq_perturbs) + list(leiden_probs) + ['total_counts']
uniq_perturbs = np.array(uniq_perturbs)
```

```python
# this ensures we have all genes rather than just the highly variable genes
adata = adata.raw.to_adata()
```

```python
# parameters for elastic net regression were from Dixit et al.
regr = ElasticNet(l1_ratio=0.5,alpha=0.0005,max_iter=10000, warm_start=False,precompute=True)

start_time = time()
coefs = run_reg(design_matrix, adata.X, adata.var.index, uniq_perturbs, regr)
print(time()-start_time)
```

```python
if os.path.isfile('all_coefs_permute.pkl'):
    if overwrite:
        run_function = True
    else:
        run_function = False
else:
    run_function = True
        
        
if run_function:
    all_coefs_permute = []
    for g_index in g_indices:
        print('gene: ',g_index)
        coefs_permute = np.zeros((coefs.shape[0],coefs.shape[1],npermutations))
        for i in range(npermutations):
            print(i)

            permuted_design = design_matrix.copy()
            indices = np.arange(design_matrix.shape[0])
            np.random.seed(i)
            np.random.shuffle(indices)

            permuted_design.iloc[:, g_index] = design_matrix.iloc[indices, g_index].values.ravel()

            start_time = time()
            coefs_permute[:,:,i] = run_reg(permuted_design,
                                          adata.X,
                                          adata.var.index,
                                          uniq_perturbs,
                                          regr).values
            print(time()-start_time)
        all_coefs_permute.append(coefs_permute)
        
    with open('all_coefs_permute.pkl','wb') as f:
        pkl.dump(all_coefs_permute,f)
else:
    with open('all_coefs_permute.pkl','rb') as f:
        all_coefs_permute = pkl.load(f)
```

```python
# get p-values from the permutations
probs = coefs.copy()
probs.loc[:,:] = 1


for g in np.arange(coefs.shape[0]):
    gene = coefs.index[g]
    for i,g_index in enumerate(g_indices):
        pert_gene = coefs.columns[g_index]

        coef = coefs.loc[gene,pert_gene]
        coefs_permute = all_coefs_permute[i]

        # if there is no coefficient set the probability to 1
        # if the coefficient is negative check the fraction of permutation coefficients that are greater than the computed coefficient
        # if the coefficient is postivie check the fraction of permutation coefficients that are less than the computed coefficient
        if (coef == 0):
            probs.loc[gene, pert_gene] = 1
        elif coef < 0:
            probs.loc[gene, pert_gene] = 1-(np.sum(coef < coefs_permute[g,g_index,:])/npermutations)
        else:
            probs.loc[gene, pert_gene] = 1-(np.sum(coef > coefs_permute[g,g_index,:])/npermutations)
```

```python
probs = probs.iloc[:,g_indices]
```

```python
# remove the columns with the cluster probabilities
cols = probs.columns
locs_to_remove = []
for i in cols:
    if 'leiden' in i:
        locs_to_remove.append(True)
    else:
        locs_to_remove.append(False)
```

```python
if np.any(locs_to_remove):
    probs = probs.iloc[:,:np.where(locs_to_remove)[0][0]]
```

```python
# do multiple testing correction only on probabilities that are not 1 (i.e. coefficient was 0 so we didn't test them)
ps = probs.values.ravel()

locs = np.where(ps != 1)
corrected_ps = multipletests(ps[locs],method='fdr_bh')[1]
ps[locs] = corrected_ps
ps = ps.reshape((probs.shape[0],
                 probs.shape[1]))
```

```python
probs.loc[:,:] = ps
```

```python
plt.hist(probs.iloc[:,0],
        bins=100)
plt.show()
```

```python
for i in range(probs.shape[1]):
    print(probs.columns[i], np.sum(probs.iloc[:,i] < 0.05))
```

```python
coefs = coefs.loc[:,probs.columns]
```

```python
coefs.shape
```

```python
# number of significant probes for each gene
sig_degs = np.sum(probs < 0.05,axis=1)
```

```python
# require at least one probe to be significant
coefs = coefs[sig_degs != 0]
```

```python
# this is the threshold for the sum of absolute value of probe coefficients 
coef_thresh = 0.1
coef_sum = np.sum(np.abs(coefs),axis=1)
```

```python
# only keep probes that are over the coefficient threshold
coef_sum = coef_sum[coef_sum > coef_thresh]
```

```python
coef_sum
```

```python
# the genes that are significantly perturbed by the probes for the
# specific gene
significantly_perturbed_genes = coef_sum.index.to_numpy().ravel()
```

```python
deg_checker = {}
for tup in degs:
    deg_checker[tup] = 0
```

Iterate through the network one step away from the perturbed gene at a time and compute the number of degs found, number of other genes found

```python
results_df = pd.DataFrame(columns = ['Depth','DEGs_found','Other_Genes', 'ndegs','ngenes'])
network_depth = 20
downstream = edges[edges.source == gene]
degs_found = pd.DataFrame(columns = ['genes', 'depth'])
other_genes = pd.DataFrame(columns = ['genes', 'depth'])
genes_already_checked = []
for depth in range(network_depth):
    print(depth)
    genes_to_check_next_round = []
    for tup in downstream.itertuples():
        if str(tup[2]) not in genes_already_checked:
            genes_to_check_next_round.append(tup[2])
            if str(tup[2]) in deg_checker:
                #print('found: ',tup[2])
                degs_found.loc[degs_found.shape[0],:] = [tup[2], depth+1]
            else:
                #print('extra:',tup[2])
                other_genes.loc[other_genes.shape[0],:] = [tup[2], depth+1]
        genes_already_checked.append(str(tup[2]))
        
    print('ngenes to check: ',len(genes_to_check_next_round))

    genes_to_check_next_round = np.unique(genes_to_check_next_round)
    
    downstream = edges[np.isin(edges.source, genes_to_check_next_round)]

        
    found_degs_num = len(np.unique(degs_found.genes))
    other_genes_num = len(np.unique(other_genes.genes))
    
    print(found_degs_num, other_genes_num, nall_degs, ngenes)
        
    results_df.loc[results_df.shape[0],:] = [depth+1, found_degs_num,
                                             other_genes_num, nall_degs,
                                             ngenes]
    
    # break out of the loop
    if len(np.unique(degs_found.genes)) + len(np.unique(other_genes.genes)) == ngenes:
        break
            
        
        
    
```

```python
# get the fraction of genes found at each
all_results = []
counter = 0


final_results_df = pd.DataFrame(columns = ['gene',
                                     'fraction_of_genes',
                                     'actual_gene_fraction',
                                     'depth',
                                     'ndegs',
                                     'ngenes',
                                    'nfound',
                                    'nothers'])




nall_degs = results_df.ndegs[0]

ngenes = results_df.ngenes[0]
frac = 0.001
for fraction_of_genes_detected in np.arange(frac,1+frac,frac):
    for tup in results_df.itertuples():
        nfound = tup[2]
        nother = tup[3]
        depth = tup[1]
        current_ngenes = nfound + nother
        current_gene_perc = current_ngenes/ngenes
        if current_gene_perc > fraction_of_genes_detected:
            break

    final_results_df.loc[final_results_df.shape[0],:] = [gene, fraction_of_genes_detected, current_gene_perc,
                                             depth, nall_degs, ngenes, nfound, nother]


```

```python
final_results_df
```

```python
# get the true positive rate and false positive rate at each depth in the network to compute the AUROC
results_df_auc = pd.DataFrame(columns = [ 'gene', 'auc_roc', 'significiant_tpr'])
counter = 0
for tup in final_results_df.groupby([ 'gene']):
    tp = tup[1].nfound
    fn = tup[1].ndegs - tup[1].nfound
    fp = tup[1].nothers
    tn = tup[1].ngenes - tup[1].ndegs - tup[1].nothers    
    
    tpr = tp / (tp + fn)
    fpr = fp / (fp + tn)
    
    fpr = np.array([0] + list(fpr) + [1])
    tpr = np.array([0] + list(tpr) + [1])
    
    sig_tpr = tpr[np.where(fpr < 0.05)[0][-1]]
    
    auc_roc = auc(fpr, tpr)
    
    #print(list(tup[0]) + [actual_auc, expected_frac_auc])
    results_df_auc.loc[results_df_auc.shape[0],:] = [tup[0],auc_roc, sig_tpr]
    
    
# plot the TPR vs FPR with the AUROC in the title
fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(fpr,tpr,s=6)
ax.plot([[0,0],[1,1]])
ax.set_xlabel('FPR')
ax.set_ylabel('TPR')
ax.set_title('AUROC: '+str(np.round(auc_roc,3)))

```
